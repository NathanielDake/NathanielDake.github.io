````python
from keras.models import Sequential 
from keras.layers import Dense # A single layer
````

````python
Sequential?
````

````python
# Here, 784 is the input to the first layer, 32 is the output (number of nodes in hidden
# layer). Hidden layer of 32 nodes maps to output layer of 10 nodes.

model = Sequential([
    Dense(32, activation="relu", input_shape=(784,)),
    Dense(10, activation="softmax"),
])

model.summary()
````

We get out total number of trainable params via:

```
784*32 + 32 + (32 * 10) + 10
```
Where the single numbers are just biases. Sequential is kind of the standard neural model. We can also make models in following fashion:

````python
model = Sequential()

model.add(Dense(32, activation="relu", input_shape=(784,)))
model.add(Dense(10, activation="softmax"))

model.summary()
````

We do need to specify the input shape. We then need to compile a model.

````python
# For multiclass classification
model.compile(
    optimizer="rmsprop",
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)
````

````python
# For binary classification
model.compile(
    optimizer="rmsprop",
    loss="binary_crossentropy",
    metrics=["accuracy"]
)
````

````python
# For mean square error regression problem
model.compile(
    optimizer="rmsprop",
    loss="mse"
)
````

# Training

````python
import numpy as np
data = np.arange(-100,100)
labels = data**2


model = Sequential([
    Dense(1, activation="relu", input_dim=1)
])

model.compile(
    optimizer="rmsprop",
    loss="mse",
    metrics=["accuracy"]
)
````

````python
model.fit(
    data,
    labels,
    epochs=1000
)
````

````python
model.predict(
    np.array([4])
)
````
