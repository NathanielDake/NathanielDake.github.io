```python
# Multi Index
import numpy as np
import pandas as pd

# Needed in order to index the results of function correctly - MultiIndexing
header = pd.MultiIndex.from_product([['Primary','Secondary'],
                                     ['window 1','window 2','window 3'],
                                     ['Frequency min', 'Value min']
                                    ],
                                    names=['P/S','window', 'sp'])
df = pd.DataFrame(np.random.randn(5, 12), 
                  index=['a','b','c','d','e'], 
                  columns=header)
df
```

```python
# Styling - can change color, etc
# docs: https://pandas.pydata.org/pandas-docs/stable/style.html
from IPython.display import HTML

def hover(hover_color="#ffff99"):
    return dict(selector="tr:hover",
                props=[("background-color", "%s" % hover_color)])

styles = [
    hover(),
    dict(selector="th", props=[
                               ("text-align", "center"), ("border-style", "solid")]),
    dict(selector="caption", props=[("caption-side", "bottom")])
]
html = (df.style.set_table_styles(styles)
          .set_caption("Hover to highlight."))
html
```
