```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%config InlineBackend.figure_format='retina'
```

```python
def p(t, a, b):
    return (1 - t) * a + t * b
```

```python
t = np.arange(0, 1, 0.001)
```

```python
a = np.array([[1], [2]])
b = np.array([[3], [1]])
```

```python
parr = p(t, a, b)
parr_x, parr_y = parr[0], parr[1]

plt.plot(parr_x, parr_y)
plt.scatter(*a)
plt.scatter(*b)
plt.xlim(0, 5)
plt.ylim(0, 5)
plt.show()
```

```python
def p2(t, a, b):
    return (1 - t**3 - t) * a + (t**3 + t) * b
```

```python
parr = p2(t, a, b)
parr_x, parr_y = parr[0], parr[1]

plt.plot(parr_x, parr_y)
plt.scatter(*a)
plt.scatter(*b)
plt.xlim(0, 5)
plt.ylim(0, 5)
plt.show()
```
