```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%config InlineBackend.figure_format='retina'
```

```python
colors = ['r', 'g', 'b']
vector_red = np.array([[-0.4], [0.6]])
vector_green = np.array([[.85], [-0.6]])
vector_blue = np.array([[0.05], [-2.0]])
vectors = [[vector_red, vector_green, vector_blue]]
```

```python
sns.set_style("whitegrid")
fig, ax = plt.subplots(figsize=(8,5))

for idx, v in enumerate(vectors[0]):
    ax.quiver([0], [0], [v[0]], [v[1]], angles='xy', scale_units='xy', color=[colors[idx]])

"""Create legend & labels"""
ax.legend(['$X_2$ = $2X_1$ + noise'], fontsize=20)
ax.set_xlabel('Centered $X_1$', fontsize=20)
ax.set_ylabel('Centered $X_2$', fontsize=20)

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,7))

p = np.array([[-1], [4]])
a = np.array([[5], [4]])
b = np.array([[2], [-3]])

V = np.array([p, a, b, a+b, a-p, b-p, p+a-p+b-p])
os = np.zeros(len(V))
origin = np.array([os, os]) 

plt.quiver(
    *origin, V[:,0], V[:,1], 
    color=['green','red','red', 'darkred', 'blue', 'blue', 'darkblue'], 
    angles='xy', scale_units='xy', scale=1)
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,7))

p = np.array([[-1], [4]])
a = np.array([[5], [4]])
b = np.array([[2], [-3]])

gamma = 0.5
V = np.array([
    p, a, b, 
    gamma*a + (1 - gamma)*b, 
    a-p, 
    b-p, 
    p+ gamma*(a-p) + (1 - gamma)*(b-p)
])
os = np.zeros(len(V))
origin = np.array([os, os]) 

plt.quiver(
    *origin, V[:,0], V[:,1], 
    color=['green','red','red', 'darkred', 'blue', 'blue', 'darkblue'], 
    angles='xy', scale_units='xy', scale=1)
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,7))

p = np.array([[-1], [4]])
a = np.array([[5], [4]])
b = np.array([[2], [-3]])


for gamma in np.arange(-10, 10, 0.1):
    # gamma = 0.5
    V = np.array([
        gamma*a + (1 - gamma)*b, 
        p+ gamma*(a-p) + (1 - gamma)*(b-p)
    ])
    os = np.zeros(len(V))
    origin = np.array([os, os]) 

    plt.quiver(
        *origin, V[:,0], V[:,1], 
        color=['red','blue','red', 'darkred', 'blue', 'blue', 'darkblue'], alpha=0.6,
        angles='xy', scale_units='xy', scale=1)
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.show()
```

What is very cool is that the constraint (the our coefficients must sum to 1) lead us to have a new type of span: the **affine span**. 

```python
fig, ax = plt.subplots(figsize=(8,7))

p = np.array([[-1], [4]])
a = np.array([[5], [4]])
b = np.array([[2], [-3]])

for gamma in np.arange(-10, 10, 0.1):
    V = np.array([
        p, a, b, 
        gamma*a + (1 - gamma)*b, 
        a-p, 
        b-p, 
        p+ gamma*(a-p) + (1 - gamma)*(b-p)
    ])
    os = np.zeros(len(V))
    origin = np.array([os, os]) 

    plt.quiver(
        *origin, V[:,0], V[:,1], 
        color=['green','red','red', 'darkred', 'blue', 'blue', 'darkblue'], alpha=0.5,
        angles='xy', scale_units='xy', scale=1)
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.show()
```
