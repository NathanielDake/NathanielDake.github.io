```python
import numpy as np
import scipy
import seaborn as sns
import matplotlib.pyplot as plt

%config InlineBackend.figure_format = 'retina'
```

```python
independent_standard_normal_rv_1 = np.random.normal(size=1000)
independent_standard_normal_rv_2 = np.random.normal(size=1000)
independent_standard_normal_rv_3 = np.random.normal(size=1000)
independent_standard_normal_rv_4 = np.random.normal(size=1000)
independent_standard_normal_rv_5 = np.random.normal(size=1000)
```

```python
Q = (
    independent_standard_normal_rv_1**2 + 
    independent_standard_normal_rv_2**2 + 
    independent_standard_normal_rv_3**2 +
    independent_standard_normal_rv_4**2 +
    independent_standard_normal_rv_5**2
)
```

```python
fig, ax  = plt.subplots(figsize=(8,5))
plt.hist(Q, bins=50, density=True, alpha=0.7)
x = np.arange(0,20,0.1)
plt.plot(x, scipy.stats.chi2.pdf(x, df=5), linewidth=4)
plt.show()
```
