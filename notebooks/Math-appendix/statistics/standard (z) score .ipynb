```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

%config InlineBackend.figure_format = 'retina'
```

```python
x = np.random.normal(loc=10, scale = 5, size=1000)
```

```python
x_axis = np.arange(-30, 30, 0.1)
mu = 10
sigma = 5
y = norm.pdf(x_axis, loc=mu, scale=sigma)
plt.plot(x_axis, y)
plt.title('Non normalized')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,7))

mu = mu - mu
y2 = norm.pdf(x_axis, loc=mu, scale=sigma)

plt.plot(x_axis, y)
plt.plot(x_axis, y2)
plt.title('Subtracted mean')
plt.legend(['Non normalized', 'Subtracted mean'])
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,7))
y3 = norm.pdf(x_axis, loc=mu, scale=1)

plt.plot(x_axis, y)
plt.plot(x_axis, y2)
plt.plot(x_axis, y3)
plt.title('Subtracted mean')
plt.legend(['Non normalized', 'Subtracted mean'])
plt.show()
```

```python
x = np.random.normal(loc=10, scale = 5, size=10000)
```

```python
fig, ax = plt.subplots(figsize=(10,7))

plt.hist(x, bins=100)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,7))
z = x - 10
plt.hist(x, bins=100, alpha=0.7)
plt.hist(z, bins=100, alpha = 0.7)
plt.legend(['X', f'X - $\mu$'])
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(14,9))
z1 = x - 10
z2 = (x - 10) / 5
plt.hist(x, bins=100, alpha=0.7)
plt.hist(z1, bins=100, alpha = 0.7)
plt.hist(z2, bins=100, alpha = 0.7)
plt.legend(['X', f'X - $\mu$', r'$Z = \frac{X - \mu}{\sigma}$'], fontsize=16)
plt.show()
```
