# Reactive Anger vs Family Environment

```python
plt.scatter(df.reactive_anger, df.family_environment)
plt.xlabel('Anger')
plt.ylabel('Family')
plt.show()
```

```python
correlation_coefficient, pval = pearsonr(df.reactive_anger, df.family_environment)
print('Correlation Coefficient: ', correlation_coefficient)
print('Pval: ', pval)
```
