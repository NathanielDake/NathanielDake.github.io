```python
def generate_vector(x_end, y_end, x_start=0, y_start=0):
    return [x_start], [y_start], [x_end], [y_end]


def vec_slope(v):
    return v[1] / v[0]


def get_line(slope, intercept=0):
    y_val1, y_val2 = (xlim_min * slope + intercept), (xlim_max * slope + intercept)
    return [xlim_min, xlim_max], [y_val1, y_val2]


def project(v, u):
    # Project v onto u
    return (u @ v) / (u @ u) * u


def get_line_from_two_vecs(v1, v2):
    return [v1[0], v2[0]], [v1[1], v2[1]]


def dot_product(v1, v2):
    return v1 @ v2


def magnitude(v):
    return np.linalg.norm(v)
```

```python
def get_angle(vector_1, vector_2):
    unit_vector_1 = vector_1 / np.linalg.norm(vector_1)
    unit_vector_2 = vector_2 / np.linalg.norm(vector_2)
    dot_product = np.dot(unit_vector_1, unit_vector_2)
    angle = np.arccos(dot_product)
    return angle
```

* Key idea: The dot product is negative 2 since it can be intepreted as: "how much of e1 is in the direction of r2?". So it is negative (and a scalar) since we get our direction from the vector r2. And it is negative because it is pointing in the *opposite* direction
* Remember: we are projecting e1 onto r2. This projection ends up being in the opposite direction of r2. When we then scale up this projection (i.e. this is the dot product step) we are asking "how much of e1 is in the direction of r2?" and of course this ends up being a negative amount, since they are pointing in opposite directions
* So the projection of e1 onto r2 is the first rotation
* We then scale this projection *by the dot product of e1 and r2*, which is negative 2!


### What happens?
* we project e1 onto rowspace (r1 and r2)
* we scale this projection by the correspond dot products (e1 dot r1 and e1 dot r2)
* We finally rotate these scaled projections to the location of e1 and e2


### Row Space is orthogonal to null space
* This makes sense. If a row vector takes in a vector, v, and maps it to zero, then that row vector was orthogonal to v. This means v resided in the null space. the set of all vectors v where this is true is the null space.
