```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.cm as cm
```

```python


# Define the linear function f(x1, x2) = c1*x1 + c2*x2
def f(x1, x2, c1, c2):
    return c1 * x1 + c2 * x2

# Constants c1 and c2
c1, c2 = 2, 3

# Generate a grid of x1, x2 values
x1_vals = np.linspace(-10, 10, 400)
x2_vals = np.linspace(-10, 10, 400)
x1, x2 = np.meshgrid(x1_vals, x2_vals)

# Evaluate f(x1, x2)
z = f(x1, x2, c1, c2)
```

```python
def set_equal_aspect_3d(ax):
    limits = np.array([ax.get_xlim3d(), ax.get_ylim3d(), ax.get_zlim3d()])
    centers = np.mean(limits, axis=1)
    max_range = np.ptp(limits, axis=1).max() / 2

    ax.set_xlim3d([centers[0] - max_range, centers[0] + max_range])
    ax.set_ylim3d([centers[1] - max_range, centers[1] + max_range])
    ax.set_zlim3d([centers[2] - max_range, centers[2] + max_range])
```

```python
# Plot the hyperplane with contour plot overlaid
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot the surface of the hyperplane
ax.plot_surface(x1, x2, z, cmap='viridis', edgecolor='none', alpha=0.4, zorder=-10)


# Labels and title
ax.set_xlabel('$x_1$')
ax.set_ylabel('$x_2$')
ax.set_zlabel('$f(x_1, x_2)$')
ax.set_title(r'Hyperplane for $f(x_1, x_2) = c_1x_1 + c_2x_2$' + f'\nwhere c1={c1}, c2={c2}')

ax.set_box_aspect(None, zoom=0.85)
plt.show()
```
