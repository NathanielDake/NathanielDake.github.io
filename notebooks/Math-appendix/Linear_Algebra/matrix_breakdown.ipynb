```python
e1 = np.array([1, 0])
r1 = np.array([1, 3])
```

```python
xlim_min = -0.2
xlim_max = 2.5
ylim_min = -0.2
ylim_max = 3.5
```

```python
def add_vec(vec, ax, color='black', start_zero=True):
    if start_zero:
        xstart = 0
        ystart = 0
    xend = vec[0]
    yend = vec[1]
    
    ax.quiver(xstart, ystart, xend, yend, angles='xy', scale_units='xy', scale=1, color=color)    
    return ax


def project(v, u):
    # Project v onto u
    return (u @ v) / (u @ u) * u


def vec_slope(v):
    return (v[1][1] - v[0][1]) / (v[1][0] - v[0][0])


def get_line(slope, intercept=0):
    y_val1, y_val2 = (xlim_min * slope + intercept), (xlim_max * slope + intercept)
    return [xlim_min, xlim_max], [y_val1, y_val2]
```

```python
def get_line(v1, v2):
    return [v1[0], v2[0]], [v1[1], v2[1]]
```

```python
e1_proj_onto_r1 = project(e1, r1)
```

```python
fig, ax = plt.subplots(figsize=(10,10))

# Vecs
ax = add_vec(e1, ax, color='red')
ax = add_vec(r1, ax, color='grey')
ax = add_vec(e1_proj_onto_r1, ax, color='blue')

# Span

xs, ys = get_line(e1, e1_proj_onto_r1)
plt.plot(xs, ys, linestyle="--")


ax.set_xlim(xlim_min, xlim_max)
ax.set_ylim(ylim_min, ylim_max)
plt.gca().set_aspect('equal', adjustable='box')

plt.show()
```
