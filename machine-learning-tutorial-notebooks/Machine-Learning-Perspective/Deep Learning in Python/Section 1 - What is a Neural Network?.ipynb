# Neural Networks with no math
* neural networks are similar to what we did with logistic regression, but with a more complex architecture 



* neural networks are like graphs, with nodes and edges as we can see in the picture above
* above there is 1 layer of input nodes, 1 layer of output nodes, and 1 layer of hidden nodes
* Deep neural networks generally have 1 or more layers of hidden nodes 
* the idea with this is that you have a bunch of inputs, which are numbers that get fed into the input nodes
    * these can be sensory receptors, like mechanoreceptors in fingertips, or visual receptors in eyes
* These input nodes get rescaled by a weight, and then get fed into the hidden nodes
* These weights are like the strength of a synapse in your brain 
    * stronger weights mean that there is a stronger connection from one node to another
    * in terms of neurons, that means that the signal would be sent from one neuron to another without much attenuation (reduction in strength of signal)
* Notice that every note in one layer is connected to every node in the next layer 

### Prediction:
You have an input signal, it travels along the edges, through the multiple layers, until it gets transmitted at the output. 

### Training:
We have to learn the weights on each edge. 

# Neural Networks are networks of neurons
* Another way to think about neural networks without thinking of graphs, is to remember what we learned about logistic regression
* remember that logistic units have a lot of similar properties to the biological neuron, so sometimes we just call it a neuron
* So when you put a bunch of neurons together, you get a neural network (one is shown in red)



# Learning (Backpropagation)
* in neural networks the learning part is called backpropagation
* imagine that we have more than 1 hidden layer. What this means is that the errors will get back propagated, one layer at a time
* In the image above:
    * V depends on error at Y
    * W depends on error at Z
    * Same pattern if more layers 
* the weights get updated based on this propagated error
